````python
# WRITE YOUR CODE HERE

import pandas as pd
import lancedb
from PIL import Image
from pdf2image import convert_from_path
from sentence_transformers import SentenceTransformer
from tqdm import tqdm

import os
import re
import time
import json
import base64
import logging
from io import BytesIO
from pathlib import Path
from typing import List, Dict, Any, Optional, TypedDict
from dataclasses import dataclass

from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate
from langgraph.graph import StateGraph, END

from dotenv import load_dotenv
load_dotenv()

logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s'
)
logger = logging.getLogger(__name__)

assert os.getenv("OPENAI_API_KEY"), "OPENAI_API_KEY not found in .env"


@dataclass
class SystemConfig:
    """
    Centralized configuration for all paths, model names, and hyperparameters.
    Ensures consistency across all pipeline stages.
    """
    pdf_directory: Path = Path("./pdfs")
    text_directory: Path = Path("./texts")
    database_path: str = "./legal_lancedb"
    collection_name: str = "legal_documents"
    vision_model: str = "gpt-4o"
    embed_model_id: str = "intfloat/multilingual-e5-large"
    generation_model: str = "gpt-4o-mini"
    max_chunk_words: int = 500
    overlap_words: int = 100
    render_dpi: int = 200

    def __post_init__(self):
        self.pdf_directory.mkdir(parents=True, exist_ok=True)
        self.text_directory.mkdir(parents=True, exist_ok=True)


cfg = SystemConfig()
````

````python
# WRITE YOUR CODE HERE
def strip_json_fences(raw_text: str) -> str:
    """Remove markdown code fences that LLMs often wrap around JSON output."""
    result = raw_text.strip()
    if result.startswith("```"):
        result = re.sub(r'^```(?:json)?\s*', '', result)
        result = re.sub(r'```\s*$', '', result)
    return result.strip()


# ---- Embedding ----
class TextEmbedder:
    """
    Wrapper around SentenceTransformer for E5 embedding models.
    E5 models require specific prefixes: 'passage:' for documents, 'query:' for queries.
    """

    def __init__(self, model_id: str):
        self.encoder = SentenceTransformer(model_id)

    def embed_document(self, content: str) -> List[float]:
        return self.encoder.encode(
            f"passage: {content}", convert_to_numpy=True
        ).tolist()

    def embed_documents_batch(self, contents: List[str]) -> List[List[float]]:
        prefixed = [f"passage: {t}" for t in contents]
        return self.encoder.encode(
            prefixed, convert_to_numpy=True, show_progress_bar=True
        ).tolist()

    def embed_question(self, question: str) -> List[float]:
        return self.encoder.encode(
            f"query: {question}", convert_to_numpy=True
        ).tolist()


# ---- OCR ----
class VisionOCR:
    """
    Performs OCR on image-based PDFs using GPT-4o vision capabilities.
    Handles Persian filenames by copying to a temp file with ASCII name.
    """

    def __init__(self, system_cfg: SystemConfig):
        self.cfg = system_cfg
        self.vision_llm = ChatOpenAI(
            model=system_cfg.vision_model, temperature=0, max_tokens=4096
        )

    def process_pdf(self, pdf_file: Path) -> str:
        import shutil
        import tempfile

        # Fix: copy file to temp path with ASCII name to avoid encoding issues
        temp_dir = tempfile.mkdtemp()
        temp_pdf = Path(temp_dir) / "document.pdf"
        try:
            shutil.copy2(str(pdf_file), str(temp_pdf))
            page_images = convert_from_path(str(temp_pdf), dpi=self.cfg.render_dpi)
        except Exception as exc:
            logger.error(f"Failed to render {pdf_file.name}: {exc}")
            return ""
        finally:
            # Clean up temp file
            if temp_pdf.exists():
                temp_pdf.unlink()
            Path(temp_dir).rmdir()

        extracted_pages = []
        logger.info(f"OCR processing: {pdf_file.name} ({len(page_images)} pages)")

        for page_idx, page_img in enumerate(page_images):
            try:
                buffer = BytesIO()
                page_img.save(buffer, format="PNG")
                encoded_img = base64.b64encode(buffer.getvalue()).decode("utf-8")

                request = HumanMessage(content=[
                    {
                        "type": "text",
                        "text": (
                            "Extract all Persian/Arabic text from this "
                            "legal document image verbatim. "
                            "Preserve structure and formatting."
                        )
                    },
                    {
                        "type": "image_url",
                        "image_url": {
                            "url": f"data:image/png;base64,{encoded_img}"
                        }
                    }
                ])
                result = self.vision_llm.invoke([request])
                extracted_pages.append(result.content)
                time.sleep(1)
            except Exception as exc:
                logger.error(f"OCR failed on page {page_idx + 1}: {exc}")
                continue

        return "\n\n".join(extracted_pages)


# ---- Validation ----
class OCRValidator:
    """Validates extracted text quality by checking length and Persian character ratio."""

    @staticmethod
    def is_valid(text: str, min_chars: int = 100) -> bool:
        if len(text.strip()) < min_chars:
            return False
        persian_count = len(re.findall(r'[\u0600-\u06FF]', text))
        return (persian_count / max(len(text), 1)) > 0.15


# ---- Metadata ----
class DocumentMetadataParser:
    """
    Extracts structured metadata from legal text content.
    Uses keyword-based scoring for legal domain classification.
    Filename keywords receive 3x weight boost.
    """

    LEGAL_DOMAINS = {
        'labor_law': [
            'کارگر', 'کارفرما', 'استخدام', 'اخراج',
            'مزد', 'قانون کار', 'بیمه کار'
        ],
        'tax_law': [
            'مالیات', 'ارزش افزوده', 'مالیاتی',
            'عوارض', 'درآمد مشمول'
        ],
        'criminal_law': [
            'قاچاق', 'جرم', 'مجازات', 'حبس',
            'جزا', 'کلاهبرداری'
        ],
        'registration_law': [
            'ثبت', 'املاک', 'سند', 'دفترخانه', 'اسناد رسمی'
        ],
        'civil_law': [
            'اجاره', 'موجر', 'مستأجر', 'عقد', 'قرارداد'
        ],
        'commercial_law': [
            'چک', 'سفته', 'برات', 'تجارت', 'شرکت'
        ],
        'social_security_law': [
            'تأمین اجتماعی', 'بازنشستگی', 'بیمه', 'حق بیمه'
        ],
    }

    def parse(self, text: str, file_name: str) -> Dict[str, Any]:
        # Extract title
        title_match = re.search(r'قانون\s+([^\n]+)', text)
        doc_title = (
            title_match.group(0).strip()[:200]
            if title_match
            else file_name.replace('.txt', '')
        )

        # Extract approval date
        date_match = re.search(r'مصوب[:\s]*(\d{4}/\d{1,2}/\d{1,2})', text)
        date_approved = date_match.group(1) if date_match else "unknown"

        # Classify domain
        domain = self._determine_domain(text, file_name)

        return {
            "document_title": doc_title,
            "approval_date": date_approved,
            "legal_domain": domain,
            "word_count": len(text.split()),
        }

    def _determine_domain(self, text: str, file_name: str) -> str:
        search_text = (text[:3000] + " " + file_name).lower()
        domain_scores = {}
        for domain, terms in self.LEGAL_DOMAINS.items():
            total = sum(
                3 if term in file_name.lower() else 1
                for term in terms if term in search_text
            )
            if total > 0:
                domain_scores[domain] = total
        return max(domain_scores, key=domain_scores.get) if domain_scores else "general_law"


# ---- Chunking ----
class ArticleBasedChunker:
    """
    Splits legal documents by article boundaries (ماده X).
    Falls back to sliding window chunking for sections exceeding max_words.
    Converts Persian numerals to Latin for article number extraction.
    """

    ARTICLE_RE = r'ماده\s*[۰-۹\d]+'

    def __init__(self, max_words: int = 500, overlap: int = 100):
        self.max_words = max_words
        self.overlap = overlap

    def split(self, text: str) -> List[Dict[str, Any]]:
        sections = self._find_article_sections(text)
        output_chunks = []
        for section in sections:
            if len(section['text'].strip()) < 30:
                continue
            output_chunks.extend(self._subdivide(section))
        return output_chunks

    def _find_article_sections(self, text: str) -> List[Dict]:
        hits = list(re.finditer(self.ARTICLE_RE, text))
        if not hits:
            return [{'type': 'general', 'number': 0, 'text': text}]

        parts = []
        if hits[0].start() > 0:
            parts.append({
                'type': 'header', 'number': 0,
                'text': text[:hits[0].start()]
            })

        for i, hit in enumerate(hits):
            boundary = hits[i + 1].start() if i + 1 < len(hits) else len(text)
            article_num = self._parse_number(hit.group())
            parts.append({
                'type': 'article', 'number': article_num,
                'text': text[hit.start():boundary]
            })
        return parts

    def _parse_number(self, header: str) -> int:
        normalized = header.translate(
            str.maketrans('۰۱۲۳۴۵۶۷۸۹', '0123456789')
        )
        match = re.search(r'\d+', normalized)
        return int(match.group()) if match else 0

    def _subdivide(self, section: Dict) -> List[Dict]:
        tokens = section['text'].strip().split()
        if len(tokens) <= self.max_words:
            return [{
                'text': section['text'].strip(),
                'chunk_index': 0,
                'section_type': section['type'],
                'section_number': section['number'],
                'is_complete_section': True
            }]
        fragments = []
        step = self.max_words - self.overlap
        frag_idx = 0
        for start in range(0, len(tokens), step):
            window = tokens[start:start + self.max_words]
            if len(window) < 20:
                continue
            fragments.append({
                'text': ' '.join(window),
                'chunk_index': frag_idx,
                'section_type': section['type'],
                'section_number': section['number'],
                'is_complete_section': False
            })
            frag_idx += 1
        return fragments
````

````python
cfg.pdf_directory = Path("./CA5-data")
cfg.text_directory = Path("./CA5-data")
cfg.pdf_directory.mkdir(parents=True, exist_ok=True)
cfg.text_directory.mkdir(parents=True, exist_ok=True)


# ---- Indexer ----
class LegalDocumentIndexer:
    """
    Orchestrates: read txt -> metadata -> chunk -> batch embed -> LanceDB.
    Uses batch embedding for efficiency (embed_documents_batch).
    Drops and recreates table on each run for clean state.
    """

    def __init__(self, system_cfg: SystemConfig):
        self.cfg = system_cfg
        self.db = lancedb.connect(system_cfg.database_path)
        self.embedder = TextEmbedder(system_cfg.embed_model_id)
        self.metadata_parser = DocumentMetadataParser()
        self.chunker = ArticleBasedChunker(
            system_cfg.max_chunk_words, system_cfg.overlap_words
        )

    def build_index(self, data_dir: Path):
        txt_files = list(data_dir.glob("*.txt"))
        if not txt_files:
            logger.error(f"No .txt files in {data_dir}")
            return None

        all_records = []

        for txt_file in tqdm(txt_files, desc="Processing documents"):
            try:
                text = txt_file.read_text(encoding='utf-8')
                if len(text.strip()) < 100:
                    logger.warning(f"Skipping {txt_file.name}: too short")
                    continue

                base_meta = self.metadata_parser.parse(text, txt_file.name)
                base_meta['source_file'] = txt_file.name
                base_meta['document_id'] = txt_file.stem

                chunks = self.chunker.split(text)
                logger.info(
                    f"{txt_file.name}: {len(chunks)} chunks, "
                    f"domain={base_meta['legal_domain']}"
                )

                for ch in chunks:
                    record = {
                        "text": ch['text'],
                        "chunk_id": (
                            f"{base_meta['document_id']}"
                            f"_c{ch['chunk_index']}"
                        ),
                        "chunk_index": ch['chunk_index'],
                        "section_type": ch['section_type'],
                        "section_number": ch['section_number'],
                        "is_complete_section": ch['is_complete_section'],
                        **base_meta
                    }
                    all_records.append(record)
            except Exception as e:
                logger.error(f"Error processing {txt_file.name}: {e}")

        if not all_records:
            logger.error("No records to index!")
            return None

        # Batch embed
        texts = [r['text'] for r in all_records]
        logger.info(f"Batch encoding {len(texts)} chunks...")
        vectors = self.embedder.embed_documents_batch(texts)
        for i, rec in enumerate(all_records):
            rec['vector'] = vectors[i]

        # Create table (handle both old and new LanceDB API)
        try:
            existing_tables = self.db.table_names()
        except (AttributeError, DeprecationWarning):
            existing_tables = self.db.list_tables()

        if self.cfg.collection_name in existing_tables:
            self.db.drop_table(self.cfg.collection_name)

        table = self.db.create_table(
            self.cfg.collection_name, data=all_records
        )
        logger.info(
            f"Indexed {len(all_records)} chunks "
            f"into '{self.cfg.collection_name}'"
        )
        return table


# ============ EXECUTE ============

# IMPORTANT: Update config paths to CA5-data
cfg.pdf_directory = Path("./CA5-data")
cfg.text_directory = Path("./CA5-data")
cfg.pdf_directory.mkdir(parents=True, exist_ok=True)
cfg.text_directory.mkdir(parents=True, exist_ok=True)

# Step 1: OCR (skip if txt exists)
ocr_engine = VisionOCR(cfg)
pdf_files = list(cfg.pdf_directory.glob("*.pdf"))
print(f"Found {len(pdf_files)} PDF files")

for pdf_path in pdf_files:
    txt_path = cfg.text_directory / f"{pdf_path.stem}.txt"
    if txt_path.exists():
        print(f"  [skip] {pdf_path.name}")
        continue
    text = ocr_engine.process_pdf(pdf_path)
    if OCRValidator.is_valid(text):
        txt_path.write_text(text, encoding='utf-8')
        print(f"  [done] {pdf_path.name}")
    else:
        print(f"  [fail] {pdf_path.name}: validation failed")

# Step 2: Index
indexer = LegalDocumentIndexer(cfg)
table = indexer.build_index(cfg.text_directory)

if table:
    row_count = table.count_rows()
    print(f"\nTotal rows in DB: {row_count}")
    sample = table.to_pandas().head(3)[
        ['chunk_id', 'legal_domain', 'section_number', 'text']
    ]
    sample['text'] = sample['text'].str[:120] + '...'
    print(sample)
else:
    print(f"\nNo data indexed. Make sure .txt files exist in {cfg.text_directory}")
````

یک نود با نام rewrite_query طراحی کنید که ورودی آن، query اصلی کاربر و خروجی آن، یک rewritten_query به‌عنوان نسخه‌ای بازنویسی‌شده و شفاف‌تر از سؤال باشد. وظیفه این ماژول آن است که ابهام‌های زبانی موجود در پرسش را برطرف کند، لحن را استاندارد و یکنواخت سازد و در نهایت پرسش کاربر را به فرمی تبدیل کند که برای مرحله‌ی بازیابی اطلاعات (retrieval) مناسب‌تر باشد.



نودی با نام classify_intent طراحی کنید که وظیفه آن تشخیص نوع پرسش کاربر بر اساس query دریافتی است. این نود باید تعیین کند که پرسش کاربر در کدام یک از دسته‌های «Greeting / Small Talk» (مانند سلام، احوال‌پرسی، تبریک و عبارات مشابه)، «Abusive / Insult / بی‌احترامی» یا «Law Question» (سؤال مرتبط با قانون و نیازمند مراجعه به پایگاه دانش) قرار می‌گیرد. در صورت تشخیص دسته‌ی Greeting، نود باید یک پاسخ کوتاه، محترمانه و دوستانه تولید کند. در صورت تشخیص دسته‌ی Abusive، نود موظف است پاسخی حرفه‌ای، محترمانه و بدون هرگونه توهین ارائه دهد و در آن توضیح دهد که مدل برای کمک طراحی شده است و وارد جدل یا بحث تنش‌زا نمی‌شود. اگر خروجی نود در دسته‌ی Law Question قرار گرفت، جریان پردازش باید به نود بعدی، یعنی نود «Metadata Extraction»، منتقل شود تا مراحل بعدی بازیابی اطلاعات انجام گیرد.





یک نود با نام extract_metadata طراحی کنید که ورودی آن rewritten_query و خروجی آن یک ساختار metadata_filter باشد؛ این ساختار در مرحله‌ی بازیابی اطلاعات از LanceDB برای اعمال فیلترهای مرتبط استفاده خواهد شد. در این نود، با استفاده از یک LLM، از روی rewritten_query اطلاعات ساختارمند و قابل‌استفاده به‌عنوان فیلتر استخراج می‌شود تا در قالب metadata_filter به LanceDB ارسال شده و بازیابی اسناد با دقت و ارتباط بیشتری انجام پذیرد.



نودی با نام context_retrieve طراحی کنید که با استفاده از rewritten_query در LanceDB عملیات جستجو را انجام دهد. این نود باید از ساختار metadata_filter برای محدود کردن فضای جستجو و اعمال فیلترهای موردنظر بهره بگیرد و در نهایت، تعداد مشخصی سند (chunk) مرتبط با پرسش کاربر، مثلاً به تعداد K سند (برای نمونه K = 10)، را به‌عنوان خروجی بازگرداند.



یک نود با نام rerank طراحی کنید که روی خروجی مرحله‌ی context_retrieve یک لایه‌ی بازمرتب‌سازی (Reranking) اعمال کند؛ به‌عنوان مثال می‌توان از یک LLM-based reranker یا یک معیار شباهت ثانویه برای امتیازدهی مجدد به اسناد استفاده کرد. این نود باید بر اساس مقادیر similarity یا score، تعداد محدودی از بهترین سندها (top-N، مثلاً N = 3) را برای مرحله‌ی تولید پاسخ (generation) انتخاب کند. علاوه بر این، در همین نود باید یک مرحله‌ی بررسی ارتباط (Relevance Check) انجام شود؛ به این صورت که اگر هیچ‌یک از contextهای انتخاب‌شده در top-N، بر اساس یک آستانه‌ی تعریف‌شده یا قضاوت LLM، ارتباط کافی با query نداشتند، فرآیند retrieval یک‌بار دیگر تکرار شده و تلاش مجدد برای بازیابی contextهای مرتبط‌تر انجام گیرد. دقت شود که نهایتا یک بار می بایست این مرحله انجام شود و Agent در loop بی نهایت قرار نگیرد پس این موضوع باید توسط شما کنترل شود.





نودی با نام generate_answer طراحی کنید که ورودی آن شامل rewritten_query به‌همراه contextهای نهایی انتخاب‌شده از مراحل قبل باشد و بر اساس این ورودی، وظیفه دارد یک پاسخ متنی نهایی، منسجم، دقیق و قابل‌فهم برای کاربر تولید و به‌عنوان خروجی برگرداند.


به طور کلی معماری پیشنهادی یک حالت از نحوه پیاده سازی این Agent هست و می توان نوع نگاه متفاوتی در طراحی معماری Agent داشت پس شما مجاز هستید که با پیاده سازی همه مواردی که بیان شد با معماری متفاوت عمل کنید معماری پیشنهادی شما می بایست تمامی این موارد را شامل شده باشد:

````python
# WRITE YOUR CODE HERE
# ------- Helper Function (MISSING!) -------
def clean_llm_json(raw: str) -> str:
    """
    Strips markdown code fences from LLM JSON output.
    LLMs often wrap JSON in ```json ... ```.
    This utility removes that wrapping before json.loads.
    """
    cleaned = raw.strip()
    if cleaned.startswith("```"):
        cleaned = re.sub(r'^```(?:json)?\s*', '', cleaned)
        cleaned = re.sub(r'```\s*$', '', cleaned)
    return cleaned.strip()


# ------- State Definition -------
class AgentState(TypedDict):
    original_query: str
    rewritten_query: str
    intent: str
    intent_response: str
    metadata_filter: Dict[str, Any]
    retrieved_docs: List[Dict[str, Any]]
    reranked_docs: List[Dict[str, Any]]
    final_answer: str
    retry_count: int
    node_timings: Dict[str, float]


# ------- LLM Initialization -------
primary_llm = ChatOpenAI(model=cfg.generation_model, temperature=0.1, max_tokens=2048)
rerank_llm = ChatOpenAI(model=cfg.generation_model, temperature=0.0, max_tokens=1024)

# ------- Shared Embedder (initialize once) -------
_embedder = None
def get_embedder() -> TextEmbedder:
    global _embedder
    if _embedder is None:
        _embedder = TextEmbedder(cfg.embed_model_id)
    return _embedder


# ------- Node 1: Rewrite Query -------
def rewrite_query_node(state: AgentState) -> AgentState:
    """Rewrites the user query to be clearer and more suitable for retrieval."""
    t0 = time.time()
    
    rewrite_prompt = ChatPromptTemplate.from_messages([
        ("system", """You are an expert at rewriting legal queries in Persian.
Your task: Rewrite the user's query to:
- Remove linguistic ambiguities
- Use formal and standard tone
- Use more precise legal terminology
- Make it more suitable for searching a legal database

Return only the rewritten query, without additional explanation."""),
        ("human", "Original query: {query}")
    ])
    
    chain = rewrite_prompt | primary_llm | StrOutputParser()
    rewritten = chain.invoke({"query": state["original_query"]})
    
    elapsed = time.time() - t0
    timings = state.get("node_timings", {}).copy()
    timings["rewrite_query"] = elapsed
    
    return {
        **state,
        "rewritten_query": rewritten.strip(),
        "node_timings": timings
    }


# ------- Node 2: Classify Intent -------
def classify_intent_node(state: AgentState) -> AgentState:
    """Classifies the user query into: greeting, abusive, or law_question."""
    t0 = time.time()
    
    classify_prompt = ChatPromptTemplate.from_messages([
        ("system", """You are an intent classification system.
Classify the user's query into one of three categories:

1. "greeting" - Hello, greetings, congratulations, friendly phrases
2. "abusive" - Insults, disrespect, profanity, inappropriate phrases
3. "law_question" - Legal question or related to law

Return output as JSON:
{{"intent": "greeting|abusive|law_question", "response": "appropriate response for greeting or abusive"}}

For greeting: Give a short, polite and friendly response in Persian.
For abusive: Give a professional and polite response in Persian explaining you are designed to help.
For law_question: Leave the response field empty."""),
        ("human", "Query: {query}")
    ])
    
    chain = classify_prompt | primary_llm | StrOutputParser()
    raw_output = chain.invoke({"query": state["original_query"]})
    
    try:
        cleaned = clean_llm_json(raw_output)
        parsed = json.loads(cleaned)
        intent = parsed.get("intent", "law_question")
        intent_resp = parsed.get("response", "")
    except (json.JSONDecodeError, KeyError):
        intent = "law_question"
        intent_resp = ""
    
    elapsed = time.time() - t0
    timings = state.get("node_timings", {}).copy()
    timings["classify_intent"] = elapsed
    
    return {
        **state,
        "intent": intent,
        "intent_response": intent_resp,
        "node_timings": timings
    }


# ------- Node 3: Extract Metadata -------
def extract_metadata_node(state: AgentState) -> AgentState:
    """Extracts metadata filters from the rewritten query for LanceDB filtering."""
    t0 = time.time()
    
    metadata_prompt = ChatPromptTemplate.from_messages([
        ("system", """You are an expert at extracting structured information from legal queries.
Extract the following information from the query and return as JSON:

{{
    "legal_domain": "labor_law|tax_law|criminal_law|registration_law|civil_law|commercial_law|social_security_law|general_law|null",
    "section_number": null or article number (integer),
    "keywords": ["relevant keywords"]
}}

If legal domain is not clear, use "null".
If article number is not mentioned, use null."""),
        ("human", "Query: {query}")
    ])
    
    chain = metadata_prompt | primary_llm | StrOutputParser()
    raw_output = chain.invoke({"query": state["rewritten_query"]})
    
    try:
        cleaned = clean_llm_json(raw_output)
        meta_filter = json.loads(cleaned)
    except (json.JSONDecodeError, KeyError):
        meta_filter = {"legal_domain": None, "section_number": None, "keywords": []}
    
    elapsed = time.time() - t0
    timings = state.get("node_timings", {}).copy()
    timings["extract_metadata"] = elapsed
    
    return {
        **state,
        "metadata_filter": meta_filter,
        "node_timings": timings
    }


# ------- Node 4: Context Retrieve -------
def context_retrieve_node(state: AgentState) -> AgentState:
    """Retrieves relevant documents from LanceDB using vector search + metadata filtering."""
    t0 = time.time()
    
    K = 10  # Number of documents to retrieve
    
    # Connect to LanceDB (use cfg, not config)
    db_conn = lancedb.connect(cfg.database_path)
    tbl = db_conn.open_table(cfg.collection_name)
    
    # Encode the rewritten query (use TextEmbedder, not EmbeddingGenerator)
    embedder = get_embedder()
    query_vec = embedder.embed_question(state["rewritten_query"])
    
    # Build search with metadata filtering
    search_builder = tbl.search(query_vec)
    
    meta = state.get("metadata_filter", {})
    
    # Apply domain filter if available
    domain_val = meta.get("legal_domain")
    if domain_val and domain_val != "null" and domain_val is not None:
        try:
            search_builder = search_builder.where(
                f"legal_domain = '{domain_val}'"
            )
        except Exception as e:
            logger.warning(f"Domain filter failed: {e}")
    
    # Apply section number filter if available
    sec_num = meta.get("section_number")
    if sec_num is not None and sec_num != "null":
        try:
            search_builder = search_builder.where(
                f"section_number = {int(sec_num)}"
            )
        except Exception as e:
            logger.warning(f"Section filter failed: {e}")
    
    results = search_builder.limit(K).to_pandas()
    
    retrieved = []
    for _, row in results.iterrows():
        doc = {
            "text": row["text"],
            "chunk_id": row.get("chunk_id", ""),
            "legal_domain": row.get("legal_domain", ""),
            "section_number": row.get("section_number", 0),
            "document_title": row.get("document_title", ""),
            "source_file": row.get("source_file", ""),
            "distance": float(row.get("_distance", 999)),
        }
        retrieved.append(doc)
    
    elapsed = time.time() - t0
    timings = state.get("node_timings", {}).copy()
    timings["context_retrieve"] = elapsed
    
    return {
        **state,
        "retrieved_docs": retrieved,
        "node_timings": timings
    }


# ------- Node 5: Rerank -------
def rerank_node(state: AgentState) -> AgentState:
    """Reranks retrieved documents using LLM-based scoring and checks relevance."""
    t0 = time.time()
    
    TOP_N = 3
    RELEVANCE_THRESHOLD = 0.3
    
    docs = state.get("retrieved_docs", [])
    if not docs:
        elapsed = time.time() - t0
        timings = state.get("node_timings", {}).copy()
        timings["rerank"] = elapsed
        return {**state, "reranked_docs": [], "node_timings": timings}
    
    # LLM-based reranking: score each document's relevance
    scored_docs = []
    
    for doc in docs:
        score_prompt = ChatPromptTemplate.from_messages([
            ("system", """You are a legal document relevance evaluator.
Return a number between 0 and 1 indicating how relevant the document is to the query.
Return only the number, without additional explanation.
1 = completely relevant, 0 = not relevant at all"""),
            ("human", "Query: {query}\n\nDocument: {document}")
        ])
        
        chain = score_prompt | rerank_llm | StrOutputParser()
        try:
            score_str = chain.invoke({
                "query": state["rewritten_query"],
                "document": doc["text"][:800]
            })
            match = re.search(r'[\d.]+', score_str)
            if match:
                relevance_score = float(match.group())
                relevance_score = min(max(relevance_score, 0.0), 1.0)
            else:
                relevance_score = 0.5
        except Exception:
            relevance_score = 0.5 - doc.get("distance", 0.5) * 0.1
        
        doc["relevance_score"] = relevance_score
        scored_docs.append(doc)
    
    # Sort by relevance score descending
    scored_docs.sort(key=lambda x: x["relevance_score"], reverse=True)
    
    # Select top-N
    top_docs = scored_docs[:TOP_N]
    
    # Relevance check: if all top-N docs are below threshold, flag for retry
    max_score = max(d["relevance_score"] for d in top_docs) if top_docs else 0
    
    current_retry = state.get("retry_count", 0)
    
    if max_score < RELEVANCE_THRESHOLD and current_retry < 1:
        # Retry: re-retrieve without metadata filters
        elapsed = time.time() - t0
        timings = state.get("node_timings", {}).copy()
        timings["rerank"] = elapsed
        return {
            **state,
            "reranked_docs": top_docs,
            "retry_count": current_retry + 1,
            "metadata_filter": {},  # Clear filters for retry
            "node_timings": timings
        }
    
    elapsed = time.time() - t0
    timings = state.get("node_timings", {}).copy()
    timings["rerank"] = elapsed
    
    return {
        **state,
        "reranked_docs": top_docs,
        "retry_count": current_retry,
        "node_timings": timings
    }


# ------- Node 6: Generate Answer -------
def generate_answer_node(state: AgentState) -> AgentState:
    """Generates the final answer using the rewritten query and reranked contexts."""
    t0 = time.time()
    
    contexts = state.get("reranked_docs", [])
    
    if not contexts:
        elapsed = time.time() - t0
        timings = state.get("node_timings", {}).copy()
        timings["generate_answer"] = elapsed
        return {
            **state,
            "final_answer": "متأسفانه اطلاعات مرتبطی در پایگاه داده قوانین یافت نشد. لطفاً سؤال خود را با جزئیات بیشتری مطرح کنید.",
            "node_timings": timings
        }
    
    # Build context string
    context_parts = []
    for i, doc in enumerate(contexts, 1):
        source = doc.get("source_file", "نامشخص")
        domain = doc.get("legal_domain", "نامشخص")
        sec = doc.get("section_number", "-")
        context_parts.append(
            f"--- Document {i} (Source: {source}, Domain: {domain}, Article: {sec}) ---\n{doc['text']}"
        )
    
    combined_context = "\n\n".join(context_parts)
    
    answer_prompt = ChatPromptTemplate.from_messages([
        ("system", """You are an expert Iranian law attorney.
Based on the provided legal documents, give an accurate, coherent and understandable answer to the user's question.

Rules:
- Only answer based on information in the documents
- Reference article numbers and legal sources
- If information is insufficient, say so honestly
- Answer in Persian
- Structure the answer clearly and organized"""),
        ("human", "Question: {query}\n\nRelevant documents:\n{context}")
    ])
    
    chain = answer_prompt | primary_llm | StrOutputParser()
    answer = chain.invoke({
        "query": state["rewritten_query"],
        "context": combined_context
    })
    
    elapsed = time.time() - t0
    timings = state.get("node_timings", {}).copy()
    timings["generate_answer"] = elapsed
    
    return {
        **state,
        "final_answer": answer.strip(),
        "node_timings": timings
    }


# ------- Routing Functions -------
def route_after_classify(state: AgentState) -> str:
    """Routes based on intent classification result."""
    intent = state.get("intent", "law_question")
    if intent in ("greeting", "abusive"):
        return "end_early"
    return "extract_metadata"


def route_after_rerank(state: AgentState) -> str:
    """Routes based on rerank results - retry or proceed to generate."""
    reranked = state.get("reranked_docs", [])
    retry_count = state.get("retry_count", 0)
    
    if not reranked:
        return "generate_answer"
    
    max_score = max(d.get("relevance_score", 0) for d in reranked)
    
    # If low relevance and haven't retried yet, go back to retrieve
    if max_score < 0.3 and retry_count == 1:
        return "context_retrieve"
    
    return "generate_answer"


def end_early_node(state: AgentState) -> AgentState:
    """Handles greeting/abusive responses without going through retrieval."""
    return {
        **state,
        "final_answer": state.get("intent_response", "سلام! چگونه می‌توانم کمکتان کنم؟")
    }


# ------- Build the Graph -------
workflow = StateGraph(AgentState)

# Add all nodes
workflow.add_node("rewrite_query", rewrite_query_node)
workflow.add_node("classify_intent", classify_intent_node)
workflow.add_node("extract_metadata", extract_metadata_node)
workflow.add_node("context_retrieve", context_retrieve_node)
workflow.add_node("rerank", rerank_node)
workflow.add_node("generate_answer", generate_answer_node)
workflow.add_node("end_early", end_early_node)

# Define edges
workflow.set_entry_point("rewrite_query")
workflow.add_edge("rewrite_query", "classify_intent")

# Conditional: after classify
workflow.add_conditional_edges(
    "classify_intent",
    route_after_classify,
    {
        "extract_metadata": "extract_metadata",
        "end_early": "end_early"
    }
)

workflow.add_edge("extract_metadata", "context_retrieve")
workflow.add_edge("context_retrieve", "rerank")

# Conditional: after rerank (retry or generate)
workflow.add_conditional_edges(
    "rerank",
    route_after_rerank,
    {
        "context_retrieve": "context_retrieve",
        "generate_answer": "generate_answer"
    }
)

workflow.add_edge("generate_answer", END)
workflow.add_edge("end_early", END)

# Compile
legal_agent = workflow.compile()
print("✅ Legal RAG Agent compiled successfully!")

# Visualize the graph
try:
    from IPython.display import display, Image
    graph_img = legal_agent.get_graph().draw_mermaid_png()
    display(Image(graph_img))
except Exception as e:
    print(f"Graph visualization: {e}")
    print(legal_agent.get_graph().draw_mermaid())
````

<h2 style="color: #24292e; border-bottom: 1px solid #eaecef; padding-bottom: 8px; margin-top: 30px;">
A. Execution Time Breakdown
</h2>

<blockquote style="background-color: #fff3cd; border-left: 5px solid #ffc107; padding: 15px; margin-left: 0; margin-right: 0;">
  <strong>Summary:</strong> The system was tested with 21 automatically generated questions (3 per legal document). The total average latency per query is approximately <strong>10.58 seconds</strong>.

### 1. Node Statistics
The following table summarizes the execution time statistics collected during the pipeline run.

| Node Name | Average Time (s) | Role |
| :--- | :---: | :--- |
| **Rerank** | **5.013s** | <span style="color:red; font-weight:bold;">Primary Bottleneck</span> |
| **Generate Answer** | **2.843s** | Final Response Generation (LLM) |
| **Extract Metadata** | **1.035s** | LLM-based Filter Extraction |
| **Rewrite Query** | **0.800s** | Query Refinement (LLM) |
| **Classify Intent** | **0.720s** | Routing Logic (LLM) |
| **Context Retrieve** | **0.167s** | Vector Search (LanceDB) |



### 2. Distribution Analysis
*   **Vector Search (Context Retrieve):** This is the fastest component (avg 0.167s). LanceDB performs extremely well even with metadata filtering, proving that the vector database is not a performance concern.
*   **LLM Nodes:** All nodes involving LLM calls (`rewrite`, `classify`, `extract`) take roughly 0.7s to 1.0s each. This is the inherent latency of the network request and token generation for the `gpt-4o-mini` model.
*   **Generate Answer:** Takes ~2.8s. This is expected as it involves generating a long, coherent Persian paragraph citing legal articles.

<h2 style="color: #24292e; border-bottom: 1px solid #eaecef; padding-bottom: 8px; margin-top: 30px;">
B. Bottleneck Identification
</h2>

<blockquote style="background-color: #ffeef0; border-left: 5px solid #f85149; padding: 15px; margin-left: 0; margin-right: 0;">
  <strong>CRITICAL FINDING:</strong> The <strong>Rerank Node</strong> is the definitive bottleneck, consuming roughly <strong>47%</strong> of the total execution time (5.01s out of 10.58s).
</blockquote>

### Why is Reranking so slow?
The current implementation uses an **LLM-based Reranker** (`gpt-4o-mini`).
1.  **Looping Mechanism:** The code iterates through the retrieved documents (Top-10) and asks the LLM to score *each one* individually or in batches.
2.  **Input Size:** The LLM must read the full text of the document chunk and the query to determine relevance.
3.  **Latency Stacking:** Unlike a specialized Cross-Encoder model (e.g., BAAI/bge-reranker), using a Chat Model for scoring incurs significant generation latency for every single document evaluated.


*(The purple bars representing "rerank" dominate the total height for almost all questions)*

---

<h2 style="color: #24292e; border-bottom: 1px solid #eaecef; padding-bottom: 8px; margin-top: 30px;">
C. Optimization Strategies
</h2>

<blockquote style="background-color: #e6fffa; border-left: 5px solid #009688; padding: 15px; margin-left: 0; margin-right: 0;">
  <strong>Objective:</strong> To reduce the total latency below 5 seconds while maintaining accuracy.
</blockquote>

### 1. Optimize the Reranker (Immediate Impact)
*   **Switch to Cross-Encoders:** Instead of using a Generative LLM (`gpt-4o-mini`) for scoring, use a specialized lightweight Cross-Encoder model like `monobeats/bert-base-multilingual-cased-fast-reranker`. These models run locally on GPU/CPU and can score 10 documents in <0.5 seconds, reducing the bottleneck by **90%**.
*   **Parallel Execution:** If keeping the LLM approach, send relevance requests for all 10 documents asynchronously (using `asyncio.gather`) rather than sequentially.

### 2. Node Fusion (Architectural Change)
*   **Combine Classification & Rewriting:** Currently, `Rewrite Query`, `Classify Intent`, and `Extract Metadata` are three separate LLM calls (~2.5s total).
*   **Proposal:** Create a single **"Router Node"**. One LLM prompt can simultaneously:
    1.  Rewrite the query.
    2.  Classify the intent.
    3.  Extract metadata JSON.
*   **Result:** This would reduce 3 network calls to 1, saving approximately **1.5 to 2.0 seconds**.

### 3. Streaming Responses
*   **User Experience (UX):** While `Generate Answer` takes ~3s, implementing **token streaming** would allow the user to see the text appearing immediately, making the system *feel* instant, even if the total processing time remains the same.

### 4. Caching
*   **Semantic Caching:** For frequent legal questions (e.g., "What is the minimum wage?"), cache the final answer. If a new query is semantically similar (cosine similarity > 0.95), return the cached answer immediately (0.1s latency).

````python
import os

# Disable LangSmith (causes 403 errors in Iran)
os.environ["LANGCHAIN_TRACING_V2"] = "false"
os.environ["LANGCHAIN_API_KEY"] = ""

# Try importing ragas
try:
    from ragas import evaluate
    from ragas.metrics import answer_relevancy, faithfulness
    from ragas.llms import LangchainLLMWrapper
    from ragas.embeddings import LangchainEmbeddingsWrapper
    from datasets import Dataset
    from langchain_openai import OpenAIEmbeddings
    RAGAS_AVAILABLE = True
    logger.info("✓ RAGAS imported successfully")
except ImportError as e:
    RAGAS_AVAILABLE = False
    logger.warning(f"✗ RAGAS not available: {e}")


def build_eval_dataset(
    questions: List[Dict], 
    data_dir: Path,
    agent_func,
    llm_instance
) -> Dict[str, List]:
    """
    Builds evaluation dataset by running questions through agent.
    Ground truth is generated by LLM from source document.
    """
    data = {
        "question": [], 
        "answer": [],
        "contexts": [], 
        "ground_truth": []
    }

    successful_count = 0
    failed_count = 0

    for item in tqdm(questions, desc="Building eval dataset"):
        try:
            # Run agent on question
            result = agent_func(item["question"], verbose=False)
            
            # Skip non-law questions
            if result.get("intent") != "law_question":
                logger.info(
                    f"Skipping non-law question: {item['question'][:50]}..."
                )
                continue

            # Extract answer and contexts
            answer = result.get("final_answer", "")
            contexts = [
                str(c.get("text", ""))
                for c in result.get("reranked_docs", [])
            ]
            
            if not contexts:
                contexts = ["اطلاعات مرتبطی یافت نشد."]

            # Generate ground truth from source document
            source_path = data_dir / item["source_file"]
            
            if source_path.exists():
                try:
                    doc_text = source_path.read_text(
                        encoding='utf-8'
                    )[:5000]
                except Exception as e:
                    logger.warning(f"Failed to read {source_path}: {e}")
                    doc_text = ""
            else:
                logger.warning(f"Source file not found: {source_path}")
                doc_text = ""

            # Generate ground truth using LLM
            gt_prompt = ChatPromptTemplate.from_messages([
                ("system",
                 "You are a Persian legal expert. Answer the question "
                 "precisely based on the document. If the answer is not "
                 "in the document, say 'اطلاعات کافی موجود نیست'. "
                 "Respond in Persian."),
                ("human", "Document:\n{doc}\n\nQuestion: {q}")
            ])
            gt_chain = gt_prompt | llm_instance | StrOutputParser()

            try:
                gt = gt_chain.invoke({
                    "doc": doc_text, 
                    "q": item["question"]
                })
            except Exception as e:
                logger.warning(f"Ground truth generation failed: {e}")
                gt = "پاسخ تولید نشد."

            # Add to dataset
            data["question"].append(item["question"])
            data["answer"].append(answer)
            data["contexts"].append(contexts)
            data["ground_truth"].append(gt)
            
            successful_count += 1

        except Exception as e:
            logger.error(
                f"Failed to process question '{item['question'][:50]}...': {e}"
            )
            failed_count += 1
            continue

    logger.info(
        f"Dataset built: {successful_count} successful, "
        f"{failed_count} failed"
    )
    return data


def run_ragas_evaluation(
    eval_data: Dict[str, List],
    output_path: str = "ragas_evaluation.xlsx"
):
    """
    Runs RAGAS evaluation and saves results to Excel.
    Falls back to manual metrics if RAGAS fails.
    """
    if not eval_data["question"]:
        logger.warning("No evaluation data to process")
        return

    # Create raw DataFrame (always works)
    raw_df = pd.DataFrame({
        "question": eval_data["question"],
        "answer": eval_data["answer"],
        "contexts": [
            " | ".join(c) for c in eval_data["contexts"]
        ],
        "ground_truth": eval_data["ground_truth"],
    })

    # Try RAGAS evaluation
    if RAGAS_AVAILABLE:
        try:
            # Setup LLM and Embeddings for RAGAS
            ragas_llm = LangchainLLMWrapper(ChatOpenAI(model="gpt-4o-mini", temperature=0))
            ragas_embeddings = LangchainEmbeddingsWrapper(OpenAIEmbeddings())
            
            # Prepare dataset
            ragas_data = {
                "question": eval_data["question"],
                "answer": eval_data["answer"],
                "contexts": eval_data["contexts"],
                "reference": eval_data["ground_truth"],
            }
            ragas_dataset = Dataset.from_dict(ragas_data)
            
            print("\n🔍 Running RAGAS evaluation...")
            eval_result = evaluate(
                dataset=ragas_dataset,
                metrics=[answer_relevancy, faithfulness],
                llm=ragas_llm,
                embeddings=ragas_embeddings,
            )
            
            print("\n" + "=" * 70)
            print("📊 RAGAS EVALUATION RESULTS")
            print("=" * 70)
            print(f"\n  Answer Relevancy: {eval_result['answer_relevancy']:.4f}")
            print(f"  Faithfulness:     {eval_result['faithfulness']:.4f}")
            
            # Save with RAGAS scores
            raw_df["answer_relevancy"] = eval_result.get("answer_relevancy", "N/A")
            raw_df["faithfulness"] = eval_result.get("faithfulness", "N/A")
            
        except Exception as e:
            logger.warning(f"RAGAS evaluation failed: {e}")
            logger.info("Falling back to manual metrics...")
    
    # Manual metrics (always calculated)
    print("\n" + "=" * 60)
    print("MANUAL EVALUATION METRICS")
    print("=" * 60)
    
    # Average answer length
    avg_answer_len = sum(len(a) for a in eval_data["answer"]) / max(len(eval_data["answer"]), 1)
    print(f"Average Answer Length: {avg_answer_len:.0f} characters")
    
    # Context usage rate
    contexts_used = sum(
        1 for c in eval_data["contexts"] 
        if len(c) > 1 or (c and c[0] != "اطلاعات مرتبطی یافت نشد.")
    )
    context_rate = contexts_used / max(len(eval_data["contexts"]), 1) * 100
    print(f"Context Usage Rate: {context_rate:.1f}%")
    
    # Average contexts per answer
    avg_contexts = sum(len(c) for c in eval_data["contexts"]) / max(len(eval_data["contexts"]), 1)
    print(f"Average Contexts per Answer: {avg_contexts:.1f}")
    
    print("=" * 60 + "\n")
    
    # Add manual metrics to DataFrame
    raw_df["answer_length"] = raw_df["answer"].str.len()
    raw_df["num_contexts"] = [len(c) for c in eval_data["contexts"]]
    raw_df["has_context"] = [
        "اطلاعات مرتبطی یافت نشد" not in " | ".join(c) 
        for c in eval_data["contexts"]
    ]
    
    # Save to Excel
    raw_df.to_excel(output_path, index=False, engine='openpyxl')
    logger.info(f"✓ Results saved to {output_path}")
    
    # Display summary
    print("\n📋 Per-Question Summary:")
    for i in range(min(5, len(eval_data["question"]))):
        print(f"\n  Q{i+1}: {eval_data['question'][i][:60]}...")
        print(f"  Answer (first 100 chars): {eval_data['answer'][i][:100]}...")
        print(f"  Contexts retrieved: {len(eval_data['contexts'][i])}")


# ============================================================
# EXECUTION
# ============================================================

# Build evaluation dataset using same questions from timing analysis
logger.info("Building evaluation dataset...")
eval_data = build_eval_dataset(
    questions=test_questions,
    data_dir=cfg.text_directory,
    agent_func=run_agent,
    llm_instance=primary_llm
)

logger.info(f"Evaluation dataset size: {len(eval_data['question'])}")

# Run evaluation
if eval_data["question"]:
    run_ragas_evaluation(
        eval_data=eval_data,
        output_path="evaluation_results.xlsx"
    )
else:
    logger.warning("No evaluation data collected")

print("\n✅ Evaluation complete!")
````

````python
# WRITE YOUR CODE HERE

chainlit_code = r'''
import os
import re
import time
import json
import logging
from pathlib import Path
from typing import List, Dict, Any, TypedDict

import lancedb
import chainlit as cl
from sentence_transformers import SentenceTransformer
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate
from langgraph.graph import StateGraph, END
from dotenv import load_dotenv

load_dotenv()

logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

# ============================================================
# CONFIG
# ============================================================

DB_PATH = "./lancedbv2"
TABLE_NAME = "legal_documents"
EMBEDDING_MODEL = "intfloat/multilingual-e5-large"
LLM_MODEL = "gpt-4o-mini"


# ============================================================
# STATE DEFINITION
# ============================================================

class AgentState(TypedDict):
    original_query: str
    rewritten_query: str
    intent: str
    intent_response: str
    metadata_filter: Dict[str, Any]
    retrieved_docs: List[Dict[str, Any]]
    reranked_docs: List[Dict[str, Any]]
    final_answer: str
    retry_count: int
    node_timings: Dict[str, float]


# ============================================================
# SHARED RESOURCES
# ============================================================

embedder_model = SentenceTransformer(EMBEDDING_MODEL)
main_llm = ChatOpenAI(model=LLM_MODEL, temperature=0.1, max_tokens=2048)
scorer_llm = ChatOpenAI(model=LLM_MODEL, temperature=0.0, max_tokens=512)


def encode_query(text: str) -> List[float]:
    return embedder_model.encode(
        f"query: {text}", convert_to_numpy=True
    ).tolist()


def clean_llm_json(raw: str) -> str:
    cleaned = raw.strip()
    if cleaned.startswith("```"):
        cleaned = re.sub(r'^```(?:json)?\s*', '', cleaned)
        cleaned = re.sub(r'```\s*$', '', cleaned)
    return cleaned.strip()


# ============================================================
# AGENT NODES
# ============================================================

def rewrite_query_node(state):
    t0 = time.time()
    prompt = ChatPromptTemplate.from_messages([
        ("system",
         "You are an expert at rewriting legal questions in Persian. "
         "Rewrite the question to be clearer and better suited for retrieval. "
         "Return only the rewritten question."),
        ("human", "Original question: {query}")
    ])
    chain = prompt | main_llm | StrOutputParser()
    rewritten = chain.invoke({"query": state["original_query"]})
    state["rewritten_query"] = rewritten.strip()
    state["node_timings"]["rewrite_query"] = time.time() - t0
    return state


def classify_intent_node(state):
    t0 = time.time()
    prompt = ChatPromptTemplate.from_messages([
        ("system",
         "Classify user input into: greeting, abusive, or law_question. "
         "Return JSON: {\"intent\": \"...\", \"response\": \"...\"}"),
        ("human", "User input: {query}")
    ])
    chain = prompt | main_llm | StrOutputParser()
    raw = chain.invoke({"query": state["original_query"]})
    try:
        parsed = json.loads(clean_llm_json(raw))
        state["intent"] = parsed.get("intent", "law_question")
        state["intent_response"] = parsed.get("response", "")
    except:
        state["intent"] = "law_question"
        state["intent_response"] = ""
    state["node_timings"]["classify_intent"] = time.time() - t0
    return state


def extract_metadata_node(state):
    t0 = time.time()
    prompt = ChatPromptTemplate.from_messages([
        ("system",
         "Extract legal metadata. Return JSON with legal_domain, "
         "section_number, keywords."),
        ("human", "Query: {query}")
    ])
    chain = prompt | main_llm | StrOutputParser()
    raw = chain.invoke({"query": state["rewritten_query"]})
    try:
        state["metadata_filter"] = json.loads(clean_llm_json(raw))
    except:
        state["metadata_filter"] = {}
    state["node_timings"]["extract_metadata"] = time.time() - t0
    return state


def context_retrieve_node(state):
    t0 = time.time()
    db_conn = lancedb.connect(DB_PATH)
    tbl = db_conn.open_table(TABLE_NAME)
    qvec = encode_query(state["rewritten_query"])
    search = tbl.search(qvec)
    meta = state.get("metadata_filter", {})
    domain = meta.get("legal_domain")
    if domain and domain != "null":
        try:
            search = search.where(f"legal_domain = '{domain}'")
        except:
            pass
    results = search.limit(10).to_pandas()
    state["retrieved_docs"] = [
        {
            "text": r["text"],
            "chunk_id": r.get("chunk_id", ""),
            "legal_domain": r.get("legal_domain", ""),
            "section_number": r.get("section_number", 0),
            "document_title": r.get("document_title", ""),
            "source_file": r.get("source_file", ""),
            "distance": float(r.get("_distance", 999))
        }
        for _, r in results.iterrows()
    ]
    state["node_timings"]["context_retrieve"] = time.time() - t0
    return state


def rerank_node(state):
    t0 = time.time()
    docs = state.get("retrieved_docs", [])
    if not docs:
        state["reranked_docs"] = []
        state["node_timings"]["rerank"] = time.time() - t0
        return state

    for doc in docs:
        try:
            prompt = ChatPromptTemplate.from_messages([
                ("system",
                 "Rate relevance 0-1. Return only a number."),
                ("human", "Question: {q}\n\nDocument: {d}")
            ])
            chain = prompt | scorer_llm | StrOutputParser()
            s = chain.invoke({
                "q": state["rewritten_query"],
                "d": doc["text"][:600]
            })
            score = float(re.search(r'[\d.]+', s).group())
            doc["relevance_score"] = min(max(score, 0.0), 1.0)
        except:
            doc["relevance_score"] = 0.5

    docs.sort(key=lambda x: x["relevance_score"], reverse=True)
    state["reranked_docs"] = docs[:3]
    state["node_timings"]["rerank"] = time.time() - t0
    return state


def generate_answer_node(state):
    t0 = time.time()
    contexts = state.get("reranked_docs", [])
    if not contexts:
        state["final_answer"] = (
            "متأسفانه اطلاعات مرتبطی یافت نشد."
        )
        state["node_timings"]["generate_answer"] = time.time() - t0
        return state

    ctx_parts = []
    for i, d in enumerate(contexts, 1):
        src = d.get("source_file", "unknown")
        domain = d.get("legal_domain", "unknown")
        sec = d.get("section_number", "-")
        ctx_parts.append(
            f"--- Document {i} (source: {src}, domain: {domain}, "
            f"article: {sec}) ---\n{d['text']}"
        )
    ctx_text = "\n\n".join(ctx_parts)

    prompt = ChatPromptTemplate.from_messages([
        ("system",
         "You are an Iranian legal expert. Answer based on documents. "
         "Answer in Persian. Cite article numbers."),
        ("human", "Question: {q}\n\nDocuments:\n{ctx}")
    ])
    chain = prompt | main_llm | StrOutputParser()
    state["final_answer"] = chain.invoke({
        "q": state["rewritten_query"],
        "ctx": ctx_text
    })
    state["node_timings"]["generate_answer"] = time.time() - t0
    return state


def end_early_node(state):
    state["final_answer"] = state.get(
        "intent_response",
        "سلام! چگونه می‌توانم کمکتان کنم؟"
    )
    return state


# ============================================================
# ROUTING
# ============================================================

def route_after_classify(state):
    if state.get("intent") in ("greeting", "abusive"):
        return "end_early"
    return "extract_metadata"


# ============================================================
# GRAPH BUILDER
# ============================================================

wf = StateGraph(AgentState)
wf.add_node("rewrite_query", rewrite_query_node)
wf.add_node("classify_intent", classify_intent_node)
wf.add_node("extract_metadata", extract_metadata_node)
wf.add_node("context_retrieve", context_retrieve_node)
wf.add_node("rerank", rerank_node)
wf.add_node("generate_answer", generate_answer_node)
wf.add_node("end_early", end_early_node)

wf.set_entry_point("rewrite_query")
wf.add_edge("rewrite_query", "classify_intent")
wf.add_conditional_edges(
    "classify_intent", route_after_classify,
    {"extract_metadata": "extract_metadata", "end_early": "end_early"}
)
wf.add_edge("extract_metadata", "context_retrieve")
wf.add_edge("context_retrieve", "rerank")
wf.add_edge("rerank", "generate_answer")
wf.add_edge("generate_answer", END)
wf.add_edge("end_early", END)

agent = wf.compile()


# ============================================================
# CHAINLIT HANDLERS
# ============================================================

@cl.on_chat_start
async def on_start():
    await cl.Message(
        content=(
            "سلام! 👋 من دستیار حقوقی هوشمند هستم.\n\n"
            "می‌توانید سوالات خود درباره قوانین کار، مالیات، جزا و سایر "
            "قوانین ایران را از من بپرسید.\n\n"
            "برای شروع، یک سوال حقوقی بپرسید! 📚⚖️"
        )
    ).send()
    cl.user_session.set("query_count", 0)


@cl.on_message
async def on_message(message: cl.Message):
    count = cl.user_session.get("query_count", 0)
    cl.user_session.set("query_count", count + 1)

    loading_msg = cl.Message(content="🔍 در حال پردازش سوال شما...")
    await loading_msg.send()

    try:
        initial_state = {
            "original_query": message.content,
            "rewritten_query": "",
            "intent": "",
            "intent_response": "",
            "metadata_filter": {},
            "retrieved_docs": [],
            "reranked_docs": [],
            "final_answer": "",
            "retry_count": 0,
            "node_timings": {}
        }

        result = agent.invoke(initial_state)

        if result.get("intent") in ("greeting", "abusive"):
            response = result.get("intent_response", "سلام!")
        else:
            response = result.get("final_answer", "متاسفانه پاسخی یافت نشد.")

            contexts = result.get("reranked_docs", [])
            if contexts:
                response += "\n\n---\n\n📄 **منابع مورد استفاده:**\n"
                for i, ctx in enumerate(contexts[:3], 1):
                    doc_title = ctx.get("document_title", "نامشخص")
                    section_num = ctx.get("section_number", "")
                    section_info = (
                        f" - ماده {section_num}" if section_num else ""
                    )
                    response += f"\n{i}. {doc_title}{section_info}"

        timings = result.get("node_timings", {})
        total_time = sum(timings.values())

        if timings:
            timing_details = " | ".join(
                [f"{k}: {v:.1f}s" for k, v in timings.items()]
            )
            response += f"\n\n⏱️ *زمان پردازش: {total_time:.1f} ثانیه*"
            response += f"\n*جزئیات: {timing_details}*"

        await loading_msg.remove()
        await cl.Message(content=response).send()

    except Exception as e:
        logger.error(f"Error processing query: {e}")
        await loading_msg.remove()
        await cl.Message(
            content=(
                f"❌ متاسفانه خطایی رخ داد:\n\n{str(e)}\n\n"
                "لطفاً دوباره تلاش کنید."
            )
        ).send()


@cl.on_chat_end
async def end():
    query_count = cl.user_session.get("query_count", 0)
    logger.info(f"Chat ended. Total queries: {query_count}")


if __name__ == "__main__":
    print("Chainlit app loaded. Run with: chainlit run app.py -w --port 8000")
'''

# Write to file
with open("app.py", "w", encoding="utf-8") as f:
    f.write(chainlit_code)

print("✅ app.py created successfully!")
print("\n📌 To run the Chainlit app:")
print("   chainlit run app.py -w --port 8000")
print("\n📌 Then open: http://localhost:8000")
````
